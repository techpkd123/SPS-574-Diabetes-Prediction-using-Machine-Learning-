```python
NA_val = data.isna().sum()
def na_filter(na, threshold = .4): #only select variables that passees the threshold
    col_pass = []
    for i in na.keys():
        if na[i]/data.shape[0]<threshold:
            col_pass.append(i)
    return col_pass
data = data[na_filter(NA_val)]
data.columns
```
